### Point 클래스

- Point 클래스를 정의: 2차원 공간의 한 점을 나타냄
- `Point` 클래스에는 점의 위치를 지정하는 생성자 `__init__(self, x, y)` 존재 
- `print_point(self)`는 점의 위치를 출력 
- `distance(self, other)`는 두 지점 사이의 거리를 반환


~~~ python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def print_point(self):
        print ("<", self.x, "," ,self.y, ">")

    def distance(self, other):
        return sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)
~~~

### Circle 클래스

- Circle 클래스는 2차원 공간의 원을 나타냄
- `self.center`와 `self.radius`를 멤버 변수로 갖고 있음
- `self.center`는 위에서 정의한 Point 클래스의 인스턴스

~~~ python
class Circle:
    def __init__(self, center, radius):
        self.center = Point(center[0], center[1])
        self.radius = radius

    def print_circle(self):
        print ("center:",self.center.x, self.center.y ,", radius:", self.radius)

    def in_circle(self, point):
        return self.center.distance(point) < self.radius
~~~
